### We will calculate the percentage difference between our price and competitors’ prices..

```python
df['variance1'] = ((df['unit_price'] - df['comp_1'])/df['comp_1'])*100
```

```python
df['variance1'].describe()
```

```python
df[df['variance1']==df['variance1'].max()]
```

```python
q1 = np.percentile(df['variance1'], 25)
q3 = np.percentile(df['variance1'], 75)
```

```python
df1_outliers= df[(df['variance1']<q1-1.5*(q3-q1))|(df['variance1']>q3+1.5*(q3-q1))]
df1_outliers
```

```python
df['variance2'] = ((df['unit_price'] - df['comp_2'])/df['comp_2'])*100
q1 = np.percentile(df['variance2'], 25)
q3 = np.percentile(df['variance2'], 75)
df2_outliers= df[(df['variance2']<q1-1.5*(q3-q1))|(df['variance2']>q3+1.5*(q3-q1))]
df2_outliers
```

```python
df['variance3'] = ((df['unit_price'] - df['comp_3'])/df['comp_3'])*100
q1 = np.percentile(df['variance3'], 25)
q3 = np.percentile(df['variance3'], 75)
df3_outliers= df[(df['variance3']<q1-1.5*(q3-q1))|(df['variance3']>q3+1.5*(q3-q1))]
df3_outliers
```

### Below we will calculate the base price of our company and for the competitors

```python
def top_trimmed_mean(x, q=0.7):
    cutoff = x.quantile(q)
    return x[x >= cutoff].mean()

base_prices = pd.DataFrame({'product_id': df['product_id'].unique()})

for col in ['unit_price', 'comp_1', 'comp_2', 'comp_3']:
    temp = (
        df.groupby('product_id')[col]
          .apply(top_trimmed_mean)
          .reset_index(name=f'base_{col}')
    )
    base_prices = base_prices.merge(temp, on='product_id', how='left')

base_prices
```

```python
df = df.merge(base_prices, on='product_id', how='left')
df
```
